```python
#combine all the filters
df2=df[(df["Event distance/length"].isin(['50km','50mi']))&(df["Year of event"]==2020)&
   (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df2.shape
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
#overwrite this column
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#Clean up athlete age 
#calculate the athlete age by subtracting by the year 
df2['Athlete_age'] = 2020-df2['Athlete year of birth']
```

```python
#remove h from athlete performance 
df2['Athlete performance']=df2['Athlete performance'].str.split('h').str.get(0)
```

```python
df2.head(2)
```

```python
#drop columns: Athlete country, athlete club, athlete year of birth, athlete age category 
df2= df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
df2.head(2)
```

```python
#clean up null value 
df2.isna().sum()
```

```python
df2[df2['Athlete_age'].isna()==1] #==1 True 0 False
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#check for duplicate values 
df2[df2.duplicated()==True]
```

```python
#reset index 
df2.reset_index(drop=True)
```

```python
#fix types 
df2.dtypes
```

```python
df2['Athlete_age']=df2['Athlete_age'].astype(int)
```

```python
df2["Athlete average speed"]=df2["Athlete average speed"].astype(float)
```

```python
df2.head(2)
```

```python
#rename columns 
df3=df2.rename(columns={'Year of event': 'year',
    'Event dates':'race_day',
    'Event name':'race_name',
    'Event distance/length':'race_length',
    'Event number of finishers':'race_number_of_finishers',
    'Athlete performance':'athlete_performance',
    'Athlete gender':'athlete_gender',
    'Athlete average speed':'athlete_average_speed',
    'Athlete ID': 'athlete_id',
    'Athlete_age':'athlete_age'                   
                        
})
```

```python
df3.head()
```

```python
#Reorder the columns 
df3 = df3[['race_day','race_name','race_length','race_number_of_finishers','athlete_id',
           'athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
#Difference in speed for the 50km, 50mi male to female
#df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()

df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].agg(['mean', 'median'])
```

For both genders, the mean and medium speeds are higher in the 50 km race compared to the 50mi race. Athletes tend to have a higher average speed in the 50km race compared to the 50mi race, which could be due to the shorter distance allowing for faster pacing. Male athletes show slightly higher average speeds than female athletes in both race lengths, though the difference is relatively small.

```python
#what age group is the best in the 50mi race (20 + races minimum)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>19').head(15)
```

```python
#what age group is the worst in the 50mi race (20 + races minimum)
df3.query('race_length == "50mi"').groupby('athlete_age')[
    'athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>19').head(15)
```

```python
#Season for the data -> slower in summer than in winter?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2 

#split between two decimals 
```

```python
df3['race_month']=df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' 
                                             if x > 11 else 'Fall' if x >8 else 'Summer' if x>5
                                             else 'Spring' if x>2 else 'Winter')
```

```python
df3.head(25)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',
                                                                                      ascending=False)
```

```python
df3.query('race_length=="50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
